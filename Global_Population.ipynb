```python
top_pop = df.sort_values('Population(in millions)', ascending=False).head(10)
plt.figure(figsize=(10,8))
sns.barplot(data=top_pop, x='Country', y='Population(in millions)', palette='Blues_r')
plt.title("Top 10 Countries by Total Population")
plt.ylabel("Population (millions)")
for i, val in enumerate(top_pop['Population(in millions)']):
    plt.text(i, val + 1, f"{val:.2f}", ha='center', fontweight='bold')
plt.show()
```

```python
top_youth = df.sort_values('Population Aged 0 to 14 (%)', ascending=False).head(5)
plt.figure(figsize=(10,8))
sns.barplot(data=top_youth, x='Country', y='Population Aged 0 to 14 (%)', palette='Greens_r')
plt.title("Top 5 Countries with Highest Youth Population (0â€“14%)")
plt.xticks(rotation=45)
for i, val in enumerate(top_youth['Population Aged 0 to 14 (%)']):
    plt.text(i, val + 0.5, f"{val:.1f}%", ha='center', fontweight='bold')
plt.show()
```

```python
top_elder = df.sort_values('Population Aged 60 and Over (%)', ascending=False).head(5)
plt.figure(figsize=(7,4))
sns.barplot(data=top_elder, x='Country', y='Population Aged 60 and Over (%)', palette='Oranges_r')
plt.title("Top 5 Countries with Highest Elderly Population (60+%)")
for i, val in enumerate(top_elder['Population Aged 60 and Over (%)']):
    plt.text(i, val + 0.5, f"{val:.1f}%", ha='center', fontweight='bold')
plt.show()
```

```python
top_dense = df.sort_values('Population density', ascending=False).head(5)
plt.figure(figsize=(7,4))
sns.barplot(data=top_dense, x='Country', y='Population density', palette='Purples_r')
plt.title("Top 5 Countries by Population Density")
for i, val in enumerate(top_dense['Population density']):
    plt.text(i, val + 5, f"{val:.1f}", ha='center', fontweight='bold')
plt.show()
```

```python
top_male = df.sort_values('Male Population(in millions)', ascending=False).head(5)
plt.figure(figsize=(7,4))
sns.barplot(data=top_male, x='Country', y='Male Population(in millions)', palette='Blues')
plt.title("Top 5 Countries by Male Population")
plt.ylabel("Male Population (millions)")

for i, val in enumerate(top_male['Male Population(in millions)']):
    plt.text(i, val + 1, f"{val:.2f}", ha='center', fontweight='bold')
plt.show()
```

```python
top_female = df.sort_values('Female Population(in millions)', ascending=False).head(5)
plt.figure(figsize=(7,4))
sns.barplot(data=top_female, x='Country', y='Female Population(in millions)', palette='Purples')
plt.title("Top 5 Countries by Female Population")
plt.ylabel("Female Population (millions)")

for i, val in enumerate(top_female['Female Population(in millions)']):
    plt.text(i, val + 1, f"{val:.2f}", ha='center', fontweight='bold')
plt.show()
```
